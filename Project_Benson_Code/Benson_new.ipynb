```python
# imports a library 'pandas', names it as 'pd'
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from IPython.display import Image

# enables inline plots, without it plots don't show up in the notebook
%matplotlib inline

# various options in pandas
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', 25)
pd.set_option('display.precision', 3)

df = pd.read_csv('https://raw.githubusercontent.com/qinglingni/QADG_Consulting/master/two_year_per4w2.csv')

df=df.drop(df.index[[0,1,2]])
```

```python
stationList=list(df.columns)
```

# BROOKLYN - MIDWOOD AREA

```python
values=list(df['AV H'])
values2=list(df['AV J'])
times=list(df['time'])
times2=list(str(df['time']))
#BEDFORD AV
#121 ST
#125 ST

plt.figure(figsize=(20, 4))

plt.subplot(1, 2, 1)
x=range(len(times))
y=values
plt.plot(x, y);
plt.scatter(x,y);
plt.ylabel('Entries');
plt.xlabel('Dates');
plt.title('AV H');
plt.xticks(x,times, rotation=90)

plt.subplot(1,2,2)
x=range(len(times))
y=values2
plt.plot(x, y);
plt.scatter(x,y);
plt.ylabel('Entries');
plt.xlabel('Dates');
plt.xticks(x,times, rotation=90)
plt.title('AV J');
```
