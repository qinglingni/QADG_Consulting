```python
import pandas as pd
df = pd.read_csv("/Users/qni/QADG_Consulting/two_year_clean.csv")
df = df.set_index("time")
df.head()
```

```python
%matplotlib inline
import matplotlib.pylab as plt
import numpy as np

df = df[np.abs(df-df.mean())<=(3*df.std())]
df = df[np.abs(df-df.mean())<=(3*df.std())]
df = df.interpolate()
df = df.rolling(5).mean()
```

```python
def plot_col(col):
    plt.figure(figsize=(12,3))
    plt.plot(range(len(df.index)),df[col])
    plt.xticks(range(0,len(df.index),10),df.index[::10])#,rotation=45)
    plt.ylabel("Count")
    plt.xlabel("Time")
    plt.title(col)
    return plt
```

```python
plot_col('AV H')
```
